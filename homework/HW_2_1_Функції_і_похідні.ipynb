1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 10 * x - 5
    
def df(x):
    if isinstance(x, (int, float)):  # Assume x is a single value
        return 10
    else:  # Assume x is a list or array-like
        return [10] * len(x)
```

```python
xs = np.linspace(0,10,20)

plt.figure(figsize=(10, 5))
plt.plot(xs, f(xs), label='f(x) = 10x-5', color='blue')
plt.plot(xs, df(xs), label="f'(x) = 10", color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
def f_sin(x):
    return np.sin(x)

def f_sin_derivative(x):
    return np.cos(x)

def f_ln(x):
    return np.log(x)

def f_ln_derivative(x):
    return 1 / x
```

```python
x_sin = np.linspace(-2 * np.pi, 2 * np.pi, 500)  # For sin(x) and its derivative
x_ln = np.linspace(0.1, 10, 500)  # For ln(x) and its derivative

# Plot sin(x) and its derivative
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x_sin, f_sin(x_sin), label='f(x) = sin(x)', color='blue')
plt.plot(x_sin, f_sin_derivative(x_sin), label="f'(x) = cos(x)", color='green')
plt.title("sin(x) and its derivative with Python")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Plot ln(x) and its derivative
plt.subplot(1, 2, 2)
plt.plot(x_ln, f_ln(x_ln), label='f(x) = ln(x)', color='blue')
plt.plot(x_ln, f_ln_derivative(x_ln), label="f'(x) = 1/x", color='green')
plt.title("ln(x) and its derivative with Python")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import sympy as sp
```

```python
x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.ln(x)


f_sin_derivative = sp.diff(f_sin, x)
f_ln_derivative = sp.diff(f_ln, x)

f_sin_derivative, f_ln_derivative
```

```python
# Convert symbolic functions and derivatives to numeric functions for plotting
f_sin_func = sp.lambdify(x, f_sin, 'numpy')
f_sin_derivative_func = sp.lambdify(x, f_sin_derivative, 'numpy')

f_ln_func = sp.lambdify(x, f_ln, 'numpy')
f_ln_derivative_func = sp.lambdify(x, f_ln_derivative, 'numpy')

# Define ranges for x values
x_sin = np.linspace(-2 * np.pi, 2 * np.pi, 500)  # For sin(x) and its derivative
x_ln = np.linspace(0.1, 10, 500)  # For ln(x) and its derivative

# Plot sin(x) and its derivative
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x_sin, f_sin_func(x_sin), label='f(x) = sin(x)', color='blue')
plt.plot(x_sin, f_sin_derivative_func(x_sin), label="f'(x) = cos(x)", color='green')
plt.title("sin(x) and its derivative with SymPy")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Plot ln(x) and its derivative
plt.subplot(1, 2, 2)
plt.plot(x_ln, f_ln_func(x_ln), label='f(x) = ln(x)', color='blue')
plt.plot(x_ln, f_ln_derivative_func(x_ln), label="f'(x) = 1/x", color='green')
plt.title("ln(x) and its derivative with SymPy")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.Symbol('x')
f = x**2 - 5*x + 1

f_derivative = sp.diff(f, x)

f_derivative
```

```python
f_numeric = sp.lambdify(x, f, 'numpy')
f_derivative_numeric = sp.lambdify(x, f_derivative, 'numpy')

x_values = np.linspace(0, 100)
f_values = f_numeric(x_values)
f_derivative_values = f_derivative_numeric(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_values, f_derivative_values, label="f'(x) = 2x - 5", color='green')
plt.title("Total Cost Function and Its Derivative")
plt.xlabel("x (Production Units)")
plt.ylabel("y (Cost)")
plt.legend()
plt.grid(True)

plt.show()
```
